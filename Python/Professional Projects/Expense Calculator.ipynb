```python
expenses=[]
def Expenses_calculator():
    Money_spend=float(input("How much mony you spend"))
    spend_category=input("On which thing you spend this Money")
    expense={"Money":Money_spend,"Category":spend_category}
    expenses.append(expense)

    print("Expenses added successfully")
def view_expenses():
    if expenses ==[]:
        print("No expenses added")
    for exp in expenses:
        print("Amount",":",exp["Money"],"\nCategory",":",exp["Category"])
def Total_expenses():
    Total=0
    if expenses ==[]:
        print("No Expenses added yet")
    for exp in expenses:
        Total+=exp["Money"]
    print("Total amount",":",Total)
```

```python
while True:
    print("Welcome to Expenses Calculator")
    print("1. Add Expenses")
    print("2. Check Expenses")
    print("3. Calculate Total")
    print("4. Exit")
    user=int(input("Choose any one number"))
    if user == 1:
        Expenses_calculator()
    elif user == 2:
        view_expenses()
    elif user == 3:
        Total_expenses()
    elif user == 4:
        break
    else:
        print("Enter Valid Number")
```
